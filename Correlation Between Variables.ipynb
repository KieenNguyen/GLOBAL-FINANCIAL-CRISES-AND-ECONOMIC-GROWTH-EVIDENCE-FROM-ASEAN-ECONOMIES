```python
data= pd.read_excel('Data_group3.xlsx')
data
```

```python
#Xoá dữ liệu thùa: 
for col in data.columns:
    if 'Unnamed' in col:
        del data[col]
```

```python
#Đổi kiểu dữ liệu thành float:
header = data.columns.drop(['code','year ']).to_series()
for name in header:
    data[name]=data[name].apply(float)
```
